```python
df.rename(columns={
    'work_year': 'Ano',
    'experience_level': 'Experiencia',
    'employment_type': 'Contrato',
    'job_title': 'Cargo',
    'salary': 'Salario',
    'salary_currency': 'Moeda',
    'salary_in_usd': 'Usd',
    'employee_residence': 'Residencia',
    'remote_ratio': 'Remoto',
    'company_location': 'Empresa',
    'company_size': 'Porte_Empresa'
}, inplace=True)
df.head()
```

```python
Experiencia = {'SE': 'Senior', 'MI': 'Pleno', 'EN': 'Junior', 'EX': 'Executivo'}
df['Experiencia'] = df['Experiencia'].replace(Experiencia)
df['Experiencia'].value_counts()
df.head()
```

```python
contrato = {'FT': 'Tempo_Integral', 'CT': 'Contrato', 'PT': 'Tempo_Parial', 'FL': 'Freelance'}
df['Contrato'] = df['Contrato'].replace(contrato)
df['Contrato'].value_counts()
df.head()
```

```python
porte_empresa = {'S': 'Pequena', 'M': 'Media', 'L': 'Grande'}
df['Porte_Empresa'] = df['Porte_Empresa'].replace(porte_empresa)
df['Porte_Empresa'].value_counts()
df.head()
```

```python
Remoto = {0: 'Presencial', 50: 'Hibrido', 100: 'Remoto'}
df['Remoto'] = df['Remoto'].replace(Remoto)
df['Remoto'].value_counts()
df.head()
```
